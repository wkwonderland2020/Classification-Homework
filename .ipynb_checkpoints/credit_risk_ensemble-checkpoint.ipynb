## Read the CSV and Perform Basic Data Cleaning

```python
# Using the panda's read_csv () to retrieve the data for analysis
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

```python
# We need to convert all categorical columns to numerical data for ease of analysis.  We use get_dummies() to convert the columns into 
# numerical data.  There were total 9 categorical columns needed to be converted into numerical data.

df = pd.get_dummies(df, columns=["home_ownership", "verification_status", "issue_d","pymnt_plan", "initial_list_status", "next_pymnt_d", "application_type",
                                 "hardship_flag","debt_settlement_flag"])
df.head()
```

## Split the Data into Training and Testing

```python
# Define new variable X and create the features by dropping the column "loan_status";
X = df.copy()
X.drop("loan_status", axis=1, inplace=True)

#  create new variable y and to create and to store our target
y = df["loan_status"]
```

```python
X.describe()
```

```python
# Using the value_counts() function to check the balance of our target values
y.value_counts()
```

```python
# Splitting the data into training and testing sets by creating X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    random_state =78)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# To avoid distortion in the analysis with raw data, we would scale the data with sklearn library's StandardScaler and to create an instance
from sklearn.preprocessing import StandardScaler
scaler =  StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset 
X_scaler = scaler.fit(X_train)
 
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

X_train_scaled
```

Balanced Random Forest Classifier### Balanced Random Forest Classifier

```python
# From imblearn library to import the BalancedRandomForest Classifier to ensemble the training data (X_trained; Y_trained); random state will kept at 1 for comparison
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
resample=BalancedRandomForestClassifier(n_estimators = 100, random_state=1)

# The model is fitted with training data, and the make predictions based on the X_test data 

resample.fit(X_train, y_train)
predictions = resample.predict(X_test)
Counter(predictions)
```

```python
# Using the BalancedRandomForest Classifier to calculate the balanced accuracy score based on
# the actual and predicted ROS_y_pred values in the model
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
brf.fit(X_train, y_train)
balanced_accuracy_score(y_test,predictions)
```

```python
# From SKlearn to import confusion matrix and
# to calculate the confusion matrix using the actual and predicted values in the model
from sklearn.metrics import confusion_matrix
confusion_matrix (y_test,predictions)
```

```python
# From imblearn to import imbalanced classification report and to  
# print the imbalanced classification report based on the actual and predicted values in the model
from imblearn.metrics import classification_report_imbalanced
print (classification_report_imbalanced(y_test,predictions))
```

```python
# List the features sorted in descending order by feature importance
importances =brf.feature_importances_
sorted(zip(brf.feature_importances_, X.columns),reverse=True)[:10]
```

### Easy Ensemble Classifier

```python
# From SK learn library to import Gradient Boosting Classifier to ensemble the training data (X_trained; Y_trained); random state will kept at 1 for comparison


# Using the loop to train the Classifier and fit the model with different learning rates: [0.05, 0.1, 0.25, 0.5, 0.75, 1];
#  each iteration will print the accuarcy score to see the performance of the model with different learning rte
from sklearn.ensemble import GradientBoostingClassifier
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    model = GradientBoostingClassifier(n_estimators=100,
                                      learning_rate=learning_rate,
                                      max_features=2,
                                      max_depth=3,
                                      random_state=1)
    model.fit(X_train_scaled,y_train.ravel())
    print("Learning rate: ", learning_rate)
    
    # Score the model
    print("Accuracy score (training): {0:.3f}".format(
        model.score(
            X_train_scaled, 
            y_train.ravel())))
    print("Accuracy score (validation): {0:.3f}".format(
        model.score(
            X_test_scaled, 
            y_test.ravel())))
    print()
```

```python
# Based on the results above, the accuracy score for validation is highest for model with learning rate of 0.05;
# As a result, we choose this learning rate for futher study

Gradient_model = GradientBoostingClassifier(n_estimators =100,
                                       learning_rate = 0.05,
                                       max_features =5,
                                       max_depth=3,
                                       random_state=1)

# Fit the model with the X_train_scaled and y_train data
Gradient_model.fit(X_train_scaled, y_train.ravel())    

# Score the model
print("Accuracy score (training): {0:.3f}".format(
    Gradient_model.score(
        X_train_scaled, 
        y_train)))
print("Accuracy score (validation): {0:.3f}".format(
    Gradient_model.score(
        X_test_scaled, 
        y_test)))
```

```python
# Based on the trained model to make prediction with the X_test_scaled data; 
Gradient_predictions = Gradient_model.predict(X_test_scaled)

# organize the prediction and actual in a dataframe

pd.DataFrame({"Prediction": Gradient_predictions, "Actual": y_test.ravel()}).head(20)
```

```python
# Generate accuracy score for predictions using y_test with the trained model
Gradient_model.score(X_test_scaled, y_test)
```

```python
# From SKlearn to compute the balanced accuracy score of the model based on the actual and predicted values
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test,Gradient_predictions)
```

```python
# Display the confusion matrix
cm = confusion_matrix(y_test, Gradient_predictions)
cm_df = pd.DataFrame(cm, index = ["Actual High Risk", "Actual Low Risk"], columns =["Predicted High Risk", "Predicted Low Risk"])
display(cm_df)

#from sklearn.metrics import confusion_matrix
#confusion_matrix(y_test, Gradient_predictions)
```

```python
# From SK learn library to import classification_report_imbalanced to print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print ("Classification Report")
print (classification_report_imbalanced(y_test, Gradient_predictions))
```
